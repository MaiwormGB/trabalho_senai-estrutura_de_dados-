```python
# - Estrutura de dados, lista de produtos...
produtos = [
    {"codigo": 1, "nome": "Remédio de feridas", "preco": 59.00, "estoque": 10, "vendedor": "flower"},
    {"codigo": 2, "nome": "spray de infecções", "preco": 120.50, "estoque": 25, "vendedor": "mai" },
    {"codigo": 3, "nome": "pomada para patinhas", "preco": 79.90, "estoque": 15, "vendedor": "apple"},
    {"codigo": 4, "nome": "bandagens", "preco": 24.90, "estoque": 13, "vendedor": "gui"},
    {"codigo": 5, "nome": "vermífugo", "preco": 84.99, "estoque": 26, "vendedor": "mat"},
    {"codigo": 6, "nome": "sabonete antibacteriano", "preco": 30.00, "estoque": 28, "vendedor": "jojo"}
]


codigo = int(input("coloque o código do seu item"))
nome = input("coloque o nome do seu item")
preco = float(input("coloque o preço do seu item"))
estoque = int(input("coloque o estoque do seu item"))
vendedor = input("coloque o nome do vendedor")

novo_produto = {"codigo": codigo, "nome": nome, "preco": preco, "estoque": estoque, "vendedor": vendedor}
produtos.append(novo_produto)

print("\n--- LISTA DE PRODUTOS ---")
for produto in produtos:
    print(f"Código: {produto['codigo']}")
    print(f"Nome: {produto['nome']}")
    print(f"Preço: R${produto['preco']:.2f}")
    print(f"Estoque: {produto['estoque']} unidades")
    print(f"vendedor: {produto['vendedor']}")
    print("-" * 30)

codigo_produto = 2
nova_quantidade = 20

for produto in produtos:
    if produto["codigo"] == codigo_produto:
        produto["estoque"] = nova_quantidade
        print(f"\nEstoque do produto {produto['nome']} atualizado para {nova_quantidade}")

codigo_remover = int(input("selecione o código do produto que deseja remover"))
for i, produto in enumerate(produtos):
    if produto["codigo"] == codigo_remover:
        removido = produtos.pop(i)
        print(f"\nProduto removido: {removido['nome']}")
        break


print("\n--- LISTA ATUALIZADA ---")
for produto in produtos:
    print(f"{produto['codigo']} - {produto['nome']}: R${produto['preco']:.2f} | Estoque: {produto['estoque']}")
```
