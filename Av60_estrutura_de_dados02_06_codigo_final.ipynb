```python
cadastros = [

     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"],
     ["nome", "senha", "email", "endereco"]

]

produtos = [
    {"codigo": 1, "nome": "Remédio de feridas", "preco": 59.00, "estoque": 10, "vendedor": "flower"},
    {"codigo": 2, "nome": "spray de infecções", "preco": 120.50, "estoque": 25, "vendedor": "mai" },
    {"codigo": 3, "nome": "pomada para patinhas", "preco": 79.90, "estoque": 15, "vendedor": "apple"},
    {"codigo": 4, "nome": "bandagens", "preco": 24.90, "estoque": 13, "vendedor": "gui"},
    {"codigo": 5, "nome": "vermífugo", "preco": 84.99, "estoque": 26, "vendedor": "mat"},
    {"codigo": 6, "nome": "sabonete antibacteriano", "preco": 30.00, "estoque": 28, "vendedor": "jojo"}
]

carrinho = {}

metodos = ("cartão de crédito", "cartão de débito", "dinheiro vivo", "PIX")
pagamentos = []

logado = False

def fazer_cadastro():
    while True:  # Loop infinito até cadastrar com sucesso ou usuário desistir
        print("INICIO")
        nome = input("Digite seu nome de usuário: ")

        # Verifica se usuário já existe
        usuario_existe = any(cadastro[0] == nome for cadastro in cadastros)

        if usuario_existe:
            print("\nErro: Este nome de usuário já está cadastrado!")
            tentar_novamente = input("Deseja tentar com outro nome? (S/N): ").upper()
            if tentar_novamente != 'S':
                return None  # Sai da função sem cadastrar
            continue  # Volta para o início do loop

        # Se usuário não existe, colete os outros dados
        senha = input("Digite sua senha: ")
        email = input("Digite seu e-mail: ")
        endereco = input("Digite seu endereço: ")

        novo_cadastro = [nome, senha, email, endereco]
        cadastros.append(novo_cadastro)
        print("\nCadastro realizado com sucesso!")
        return novo_cadastro  # Sai da função com sucesso

def fazer_login():
  global logado
  while True:
    print("INICIO")
    nome = input("Digite seu nome de usuário")
    senha = input("Digite sua senha")

    usuario_existe1 = any(cadastro[0] == nome for cadastro in cadastros)
    usuario_existe2 = any(cadastro[1] == senha for cadastro in cadastros)

    if usuario_existe1 and usuario_existe2:
      print(f"Login efetuado com sucesso! bem vindo {nome}")
      logado = True
      return
    else:
      print("Erro: senha ou nome incorretos. tente novamente")

while logado == False:

  print("_" *50)
  print("\nBem vindo(a) a loja virtual Mystic™ ヾ(•ω•`)o\nA melhor opção para seu amigo de quatro patas!")
  print("_" *50)
  print("\nJa possui login?\n(Y/N)")
  escolha = input("")
  escolha = escolha.upper()

  if escolha == 'N':

    print("_" *50)
    print("Faça seu cadastro para continuar!1!!11")
    fazer_cadastro()

  elif escolha == 'Y':

    print("_" *50)
    print("Faça seu login:")
    fazer_login()

  else:
    print("Escolha invalida. tente novamente mais tarde")





codigo_produto = 2
nova_quantidade = 20

for produto in produtos:
    if produto["codigo"] == codigo_produto:
        produto["estoque"] = nova_quantidade
        print(f"\nEstoque do produto {produto['nome']} atualizado para {nova_quantidade}")

while True:
    print("_" * 50)
    print("\n--- LISTA DE PRODUTOS ---")

    for produto in produtos:
      print(f"Código: {produto['codigo']}")
      print(f"Nome: {produto['nome']}")
      print(f"Preço: R${produto['preco']:.2f}")
      print(f"Estoque: {produto['estoque']} unidades")
      print(f"vendedor: {produto['vendedor']}")
      print("-" * 30)

    escolha = input("\nDigite o código do produto (ou 0 para sair): ")

    if escolha == "0":
        break

    try:
        codigo = int(escolha)
        produto_encontrado = None

        for produto in produtos:
            if produto["codigo"] == codigo:
                produto_encontrado = produto
                break

        if produto_encontrado:
            quantidade = int(input(f"Quantidade de {produto_encontrado['nome']}: "))

            if codigo in carrinho:
                carrinho[codigo] += quantidade
                produto["estoque"] -= quantidade
                print("Seu carrinho:", carrinho)
            else:
                carrinho[codigo] = quantidade
                produto["estoque"] -= quantidade
                print("Seu carrinho:", carrinho)

            print(f"{quantidade}x {produto_encontrado['nome']} adicionado(s) ao carrinho!")
        else:
            print("Código inválido!")

    except ValueError:
        print("Por favor, digite um número válido.")


print("Métodos de pagamento:")
for i, m in enumerate(metodos, 1):
    print(f"{i} - {m}")

escolha = int(input("Escolha o número do método: "))
metodo = metodos[escolha - 1]

valor = float(input("Valor do pagamento: R$"))
promocao_possivel = input(f"Seu pagamento tem alguma promoção? \n(S ou N): ")
promocao_possivel = promocao_possivel.upper()

if promocao_possivel == "S":
  porcentagem_promocao = float(input("Quanto é? (exemplo: '10' vai ser igual a 10%):\n"))
  desconto = valor * (porcentagem_promocao / 100)
  valor_com_desconto = valor - desconto
else:
  desconto = 0
  valor_com_desconto = valor

pagamentos.append({"metodo": metodo, "valor": valor})

print("\nPagamentos registrados:")
for p in pagamentos:
    print(f"Você pagou R$", valor_com_desconto, f"via {p['metodo']}")
```
